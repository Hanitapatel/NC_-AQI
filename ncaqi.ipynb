```python
# Load the AQI dataset.
file_path ="C:/Users/tgeis/OneDrive/Desktop/Module 20 Final Proj/US_AQI.csv"
aqi_df = pd.read_csv(file_path, index_col=0)
aqi_df.head()
```

```python
# Keep NC data
ncaqi_df = aqi_df.loc[aqi_df["state_id"]== "NC"]
ncaqi_df.head()
```

```python
print(ncaqi_df["city_ascii"].unique())
```

```python
# Split date into 3 columns for day, month, year

ncaqi_df.Date = pd.to_datetime(ncaqi_df.Date)
ncaqi_df['year'],ncaqi_df['month'],ncaqi_df['day'] = ncaqi_df.Date.dt.year, ncaqi_df.Date.dt.month, ncaqi_df.Date.dt.day
ncaqi_df.head()
```

```python
ncaqi2_df =ncaqi_df.reindex(columns =['CBSA Code','Date','year','month','day','AQI','Category','Defining Parameter','Number of Sites Reporting','city_ascii','state_id','state_name','lat','lng','population','density','timezone'])
ncaqi2_df.head()
```
